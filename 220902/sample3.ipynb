```python
# 지정된 경로에 있는 파일 목록

files = os.listdir('../csv/2017')

files
```

```python
files
```

```python
# 1. 파일 목록 출력
# 2. 파일 목록을 불러와서 concat() 하나의 데이터프레임으로 결합

df_2017 = pd.DataFrame()

path = '../csv/2017/'

for file in files :
    df = pd.read_csv(path + file)
    df_2017 = pd.concat([df_2017, df], axis = 0, ignore_index = True)

df_2017
```

```python
# 1. csv/2018에 있는 파일들도 df_2018로 데이터프레임 변환
# 2. total_df를 만들어서 2017 데이터와 2018 데이터를 결합

pd.read_csv('../csv/2018/201801_expense_list.csv')
```

```python
files2 = os.listdir('../csv/2018')

files2
```

```python
df_2018 = pd.DataFrame()

path2 = '../csv/2018/'

for file2 in files2 :
    df2 = pd.read_csv(path2 + file2)
    df_2018 = pd.concat([df_2018, df2], axis = 0, ignore_index = True)

df_2018
```

```python
# 2. total_df를 만들어서 2017 데이터와 2018 데이터를 결합

# total_df = df_2017과 df_2018을 행 추가로 단순 결합

total_df = pd.concat([df_2017, df_2018], axis = 0, ignore_index = True)

total_df
```

```python
# 1. 컬럼의 이름 변경['nid', '제목', 'url', '부서레벨1', '부서레벨2', '부서레벨3', '부서레벨4', '부서레벨5',
#                 '집행연도', '집행월', '예산', '집행', '구분', '부서명', '집행일시', '집행장소', '집행목적',
#                 '대상인원', '결제방법', '집행금액']

total_df.columns = ['nid', '제목', 'url', '부서레벨1', '부서레벨2', '부서레벨3', '부서레벨4', '부서레벨5',
                    '집행연도', '집행월', '예산', '집행', '구분', '부서명', '집행일시', '집행장소', '집행목적',
                    '대상인원', '결제방법', '집행금액']

total_df.head()
```

```python
# 2. 컬럼 삭제 → '부서레벨3', '부서레벨4', '부서레벨5', '예산', '집행', '구분'

total_df = total_df.dropna(axis = 'columns')

total_df
```

```python
# 3. 연도별 집행금액의 합 → 시각화(bar)

month_total = total_df['집행연도'].unique()       # 칼럼의 종류 확인
```

```python
exec_sum = total_df.groupby('집행연도')['집행금액'].sum()

exec_sum
```

```python
import matplotlib.pyplot as plt
import matplotlib
```

```python
# bar 그래프
# x축: index
# y축: 집행금액의 합계

plt.bar(month_total, exec_sum)
plt.show()
```

```python
# pivot_table() 형태로 시각화 변경

month_total = pd.pivot_table(total_df, index = ['집행연도'], values = '집행금액', aggfunc = sum)

month_total
```

```python
from matplotlib import font_manager, rc
```

```python
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

(month_total.plot.bar())
```

```python
# 연도별 월별

month_total = total_df.groupby(['집행연도', '집행월']).sum()['집행금액']

month_total
```

```python
month_total.plot.bar()
```

```python
pivot_month = pd.pivot_table(total_df, index = ['집행월'], columns = ['집행연도'], values = ['집행금액'], aggfunc = 'sum')

pivot_month
```

```python
pivot_month.plot.bar()
```

```python
# 집행연도와 부서레벨1 별 집행 금액의 평균이 어떠한가 그래프로 표시

month_level = total_df.groupby(['집행연도', '부서레벨1']).mean()['집행금액']

month_level
```
