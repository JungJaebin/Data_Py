```python
import matplotlib.pyplot as plt
```

```python
# 라인 그래프 출력

plt.plot([1, 2, 3, 4])
plt.show()
```

```python
# 라인 그래프 x축의 값, y축의 값 지정

plt.plot([1, 2, 3, 4], [1, 4, 2, 8])
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [7, 23, 6, 19]

plt.plot(x, y, "--r")           # '--'는 점선 그래프, 'r'은 빨간색
plt.plot(x, [4, 14, 19, 20], "oy")                #'o'는 점으로 표시, 'y'는 노란색
plt.xlabel("X")             # x축의 이름
plt.ylabel("Y")             # y축의 이름
plt.title("X and Y")        # 그래프의 제목
plt.legend(["A", "B"], loc = "best")      # legend()는 선의 제목 및 종류 표시 , loc는 위치 표시
plt.show()
```

```python
# subplot

plt.subplot(1, 2, 1)                # subplot(a, b, c): a는 행, b는 열, c는 위치
plt.plot([1, 2, 3, 4])

plt.subplot(1, 2, 2)
plt.plot([5, 31, 18, 25])

plt.show()
```

```python
# 임의의 그래프 4개를 2x2 subplot 출력

plt.subplot(2, 2, 1)
plt.plot([1, 2, 3, 4])

plt.subplot(2, 2, 2)
plt.plot([5, 31, 18, 25])

plt.subplot(2, 2, 3)
plt.plot([10, 1, 1, 7])

plt.subplot(2, 2, 4)
plt.plot([2, 9, -2, 6])

plt.show()
```

```python
# 임의의 그래프 4개를 2x2 subplot 출력

plt.subplot(3, 3, 1)
plt.plot([1, 2, 3, 4])

plt.subplot(3, 3, 2)
plt.plot([5, 31, 18, 25])

plt.subplot(3, 3, 3)
plt.plot([10, 1, 1, 7])

plt.subplot(3, 3, 4)
plt.plot([2, 9, -2, 6])

plt.subplot(3, 3, 5)
plt.plot([4, 3, 2, 1])

plt.subplot(3, 3, 6)
plt.bar([2, 9, -2, 6], [5, 2, 12, 6])

plt.show()
```

```python
# 막대 그래프

x = [1, 2, 3, 4]
y = [21, 6, 14, 5]

plt.bar(x, y)
plt.show()
```

```python
plt.bar(x, y, color = ["r", "b"], width = 0.5)          # 색깔은 입력한 순서대로
plt.show()
```

```python
plt.bar(x, y, color = ["r", "b", "c"], width = 0.5)
plt.xticks(x, ["maths", "science", "history", "english"])
plt.show()
```

```python
import numpy as np
```

```python
# 산점도 그래프

x = np.random.rand(50)
y = np.random.rand(50)

print(x)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
# 산점도 그래프 옵션 추가(영역 확장, 색상 변경, 투명도 조절)

color = np.random.rand(50)
area = (20 * np.random.rand(50)) ** 2               # random.rand의 숫자가 0 ~ 1까지 이다 보니 범위가 너무 작아서 표시 범위를 키워주기 위해 20*, **2

plt.scatter(x, y, s = area, c = color, alpha = 0.5)             # alpha: 0 ~ 1 까지의 투명도
plt.show()
```
